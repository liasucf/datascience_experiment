```python
import pandas as pd
from sklearn.model_selection import train_test_split 
#pour l'entra√Ænement automatique:
from supervised.automl import AutoML
```

```python
data =pd.read_csv("../data/raw/mars-2014-complete.csv" ,encoding='latin-1', sep=";")
```

```python
data.columns
```

```python
data = data.drop(columns=['date_maj','Unnamed: 26', 'hc', 
       'Unnamed: 27', 'Unnamed: 28', 'Unnamed: 29', 'hcnox', 'date_maj' ])
```

```python
data = data[data['co2'].isna()==False] 
```

```python
y = data['co2'].values
X = data.drop(columns=['co2'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
automl = AutoML(total_time_limit=5*60,mode='Explain',random_state=42,ml_task='regression')
#fit models
automl.fit(X_train,y_train)
#predictions
predictions = automl.predict(X_test)
#generate html report
automl.report()
```
